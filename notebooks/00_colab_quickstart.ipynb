```python
import numpy as np
from qatne.algorithms.qatne_solver import QATNESolver

hamiltonian = np.eye(4)
solver = QATNESolver(hamiltonian=hamiltonian, num_qubits=2, shots=256)
params = np.zeros(solver._estimate_num_parameters())
energy = solver._compute_energy(params)
print(f'✅ Colab smoke test energy: {energy:.4f}')
```
